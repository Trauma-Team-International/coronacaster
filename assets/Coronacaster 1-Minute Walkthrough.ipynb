## Get Data

A dataset containg daily cases and deaths. The source of the data is [European Union](https://data.europa.eu/euodp/en/data/dataset/covid-19-coronavirus-data/resource/55e8f966-d5c8-438e-85bc-c7a5a26f4863).

```python
data = coronacaster.get_data_from_eu()
```

<hr>

## Perform the Experiment

```python
results = coronacaster.forecast('Finland', data, startdate='2020-03-20', ftype='poly3')
```
